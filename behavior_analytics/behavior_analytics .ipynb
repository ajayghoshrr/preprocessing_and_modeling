```python
x['Monetory'] = x.Quantity*x.UnitPrice
x.head(1)
```

```python
x = x[x.Quantity>0]
x['Frequency'] = x.groupby('CustomerID').CustomerID.transform(lambda x:len(x))
x.head(3)
```

```python
from datetime import datetime as dt
now = dt(2011,12,11)
now
x.InvoiceDate.dtype
```

```python
#converted the content into datetime format
x.InvoiceDate = pd.to_datetime(x.InvoiceDate)
x.head(3)
```

```python
x['Recency'] = x.groupby('CustomerID').InvoiceDate.transform(lambda x: now - max(x))
```

```python
x.head(2)
```

```python
x['Monetory'] = x.groupby('CustomerID').Monetory.transform(lambda x:x.sum())
```

```python
data2 = x[['CustomerID','Monetory','Frequency','Recency']]
data2.drop_duplicates(inplace=True)
```

```python
data2.head(3)
summary = data2.describe()
summary
```

```python
def quartile(x):
    if x<18:
        return 1
    elif x<51 and x>18:
        return 2
    elif x>51 and x<141:
        return 3
    else:
        return 4

    
def quartile1(x):
    if x>1664:
        return 1
    elif x<1664 and x>674:
        return 2
    elif x>307 and x<674:
        return 3
    else:
        return 4


def quartile2(x):
    if x>100:
        return 1
    elif x<100 and x>41:
        return 2
    elif x>17 and x<41:
        return 3
    else:
        return 4


data2['RQ']= x.Recency.transform(lambda x:quartile(x.days))
data2['MQ']= x.Monetory.transform(lambda x:quartile1(x))
data2['FQ']= x.Frequency.transform(lambda x:quartile2(x))

data2[data2.RQ==1][data2.MQ==1][data2.FQ==1].count() #111

```
